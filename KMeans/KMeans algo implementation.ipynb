```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
x,y = make_blobs(n_samples=300, n_features=2,centers=4, cluster_std=1.8,random_state=101)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
wcv = []
for i in range(1,16):
    km = KMeans(n_clusters=i)
    km.fit(x)
    wcv.append(km.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,16),wcv)
plt.show()   
```

```python
km = KMeans(n_clusters=4)
y_pred = km.fit_predict(x)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[y_pred == 0,0],x[y_pred == 0,1], c="r", )
plt.scatter(x[y_pred == 1,0],x[y_pred == 1,1], c="g")
plt.scatter(x[y_pred == 2,0],x[y_pred == 2,1], c="y")
plt.scatter(x[y_pred == 3,0],x[y_pred == 3,1], c="b")
plt.scatter(x[y_pred == 4,0],x[y_pred == 4,1], c="magenta")
plt.xlabel('salary')
plt.ylabel('expenses')
plt.title('Customer Classification')
```

```python
data = make_blobs(n_samples=300, n_features=2,centers=4, cluster_std=1.8,random_state=101)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
km = KMeans(n_clusters=10)
```

```python
km.fit(data[0])
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
plt.figure(figsize=(20,20))
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means predicted')
ax1.scatter(data[0][:,0],data[0][:,1],c=km.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
